# DATA VISUALIZATION 

```python
import numpy as np
import pandas as pd

# Import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].isna().sum()
```

```python
df = df.dropna(subset=["TotalCharges"])
df.info()
```

```python
df["Churn"].value_counts(normalize=True)
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
df.info()
```

```python
plt.figure(figsize = (15,5))
sns.histplot(data=df,x="TotalCharges")
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.histplot(data=df,x="MonthlyCharges")
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.histplot(data=df,x="tenure")
plt.show()
```

```python
sns.boxplot(data=df,x='MonthlyCharges') 
plt.show()
```

```python
df["Partner"] = df["Partner"].map({"Yes": 1, "No": 0})

churn_by_partner = (
    df.groupby("Partner")["Churn"]
      .mean()
      .reset_index()
)

sns.barplot(x="Partner", y="Churn", data=churn_by_partner)
plt.xticks([0, 1], ["No Partner", "Has Partner"])
plt.ylabel("Churn Rate")
plt.title("Churn Rate by Partner Status")
plt.show()
```

```python
df["Dependents"] = df["Dependents"].map({"Yes": 1, "No": 0})

churn_by_dependents = (
    df.groupby("Dependents")["Churn"]
      .mean()
      .reset_index()
)

sns.barplot(x="Dependents", y="Churn", data=churn_by_dependents)
plt.xticks([0, 1], ["No Dependents", "Has Dependents"])
plt.ylabel("Churn Rate")
plt.title("Churn Rate by Dependents Status")
plt.show()
```

```python

df["PhoneService"] = df["PhoneService"].map({"Yes": 1, "No": 0})

churn_by_phoneservice = (
    df.groupby("PhoneService")["Churn"]
      .mean()
      .reset_index()
)

sns.barplot(x="PhoneService", y="Churn", data=churn_by_phoneservice)
plt.xticks([0, 1], ["No PhoneService", "Has PhoneService"])
plt.ylabel("Churn Rate")
plt.title("Churn Rate by PhoneService Status")
plt.show()
```
